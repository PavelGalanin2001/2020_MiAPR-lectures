# Адаптивный шаг обучения

Рассмотрим последовательное обучение. Для получения адаптивного шага будем использовать метод наискорейшего спуска.

$y(t) = x^2(t)$

$x(t+1) = x(t) - \alpha {\partial y(t) \over \partial x(t)}$

$x(t+1) = x(t) - \alpha 2 x(t)$

$y(t+1) = x^2(t)$

$\alpha=\min(y(t+a))$

<hr />

$y(t+1) = x^2(t+1) = (x(t) - \alpha 2 x(t))^2$

${\partial y(t+1) \ \partial \alpha} = 2~(x(t) - \alpha~2~x(t))~(-2~x(t)) = 0$

$x(t) - \alpha~2~x(t) = 0 \Rightarrow \alpha = {1 \over 2}$

${\partial^2 y(t+1) \over \partial \alpha^2} = 8 x^2(t) > 0$ - 2-ая произодная

$\alpha = \min$ - точка минимума

<hr />

$x(t+1) = x(t) - {1 \over 2}~2~x(t) = 0$

Рассмотрим линейный персептрон

![img](imgs/3.png)

$y_i(t) = \sum\limits_{n=1}^n \omega_{ij}(t) x_i(t) - T_j(t)$

$\omega_j(t+1) = \omega_{ij}(t) - \alpha~{\partial E \over \partial \omega_{ij}(t)}$

$T_j(t+1) = T_j(t) - \alpha {\partial E \over \partial T_j(t)}$

$E = {1\over2}~\sum\limits_{j=1}^m (y_i(t) - e_j)^2$

$y_j(t+1) = \sum\limits_{n=1}^n \omega_j(t+1) x_i(t) - T_j(t)$

![img](imgs/4.png)

$y_j(t+1) = \sum\limits_{n=1}^n x_i(t) \omega_{ij}(t)$

$\omega_{ij}(t+1) = \omega_{ij}(t) - \alpha~{\partial E \over \partial \omega_{ij}(t)}$

$T_j(t+1) = T_j(t) - \alpha~{\partial E \over \partial T_j(t)}$

$y_j(t+1) = \sum\limits_{i=1}^n x_i(t) \omega_{ij}(t) - T_j(t) - \alpha~(\sum\limits_{i=1}^n {\partial E \over \partial \omega_j(t)}) x_i(t) - {\partial E \over \partial T_j(t)}
= y_j(t) - \alpha b_j$

$y_j(t+1) = y_j(t) - \alpha b_j$

$E' = {1\over2} \sum_{j=1}^m (y_j(t+1) - e_j)^2 = {1\over2}\sum\limits_{i=1}^m (y_i(t) - \alpha b_j - e_j)^2$

$\alpha = \min(E)$

${\partial E' \over \partial\alpha} = \sum\limits_{j=1}^n (y_j(t) - \alpha b_j - e_j)~(-b_j) = 0$

$\alpha \sum\limits_{j=1}^m b_j^2 = \sum\limits_{j=1}^m (y_j(t) - e_j) b_j$

$\alpha = {\sum\limits_{j=1}^m (y_j(t) - e_j) b_j \over \sum\limits_{j=1}^m b_j^2}$

${\partial^2 E' \over \partial \alpha^2} = \sum\limits_{j=1}^m b_j^2 > 0$ - 2-ая производная

$\alpha = \min$

$b_j = \sum\limits_{i=1}^n {\partial E \over \partial \omega_{ij}(t)} x_i(t) - {\partial E \over \partial T_j(t)}$

${\partial E \over \partial w_{ij}(t)} = x_i(t) ~ (y_i(t) - e_j)$

${\partial E \over T_j(t)} = -(y_i(t) - e_j)$

$b_j = \sum\limits_{n=1}^n (y_j(t) - e_j)~x_i^2(t) + (y_1(t) - e_j)$

$b_j = (y_j(t) - e_j)~(1 + \sum\limits_{i=1}^n x_i^2(t))$

$\alpha = {
\sum\limits_{i=1}^m (y_j(t) - e_j)^2~(1 + \sum\limits_{i=1}^n x_i^2(t)) 
\over
\sum\limits_{j=1}^n (y_j(t) - e_j)^2~(1 + \sum\limits_{i=1}^n x_i^2)^2
}
= {1 \over 1 + \sum\limits_{i=1}^n x_i^2}$

$\alpha = {1 \over 1 + \sum\limits_{i=1}^n x_i^2}$

**Примичание**

Если при выводе выражения для $\alpha$ не использовать пороговое значение нейроный элементов, то величина адаптивного шага будет изменяться следующим образом

$\alpha = {1 \over \sum\limits_{i=1}^n x_i^2}$

$y_{ij}(t) = \sum\limits_{i=1}^n \omega_{ij}(t) x_i(t)$
